## Handling the missing data in the training set

we will impute mean values in the age column and will remove the Cabin column since it contains a large number of missing values as compared to the count of rows in the training data. For the missing the data in the embarked column we fill remove the 2 rows with the missing data.

```python
train_data['Cabin'].unique()
```

```python
train_data['Embarked'].unique()
```

```python
train_data['Age'].fillna(train_data['Age'].mean(), inplace=True)
```

```python
train_data = train_data.drop('Cabin', axis=1)
```

```python
train_data = train_data.dropna()
```

```python
train_data.isnull().sum()
```

## Preparing data for modelling (PreProcessing)

There are some columns in the data that are of no use for our machine learning model, hence we will get rid of those column in this section.

Also we will be 
1. converting Non-Numeric data into Numeric
2. Label Encoding if required
3. Scale the feature values to a uniform range
4. Split the data into train and test sets

```python
train_data = train_data.drop(['PassengerId', 'Name'], axis=1)
```

```python
train_data
```

```python
from sklearn.preprocessing import LabelEncoder

la = LabelEncoder()

for col in train_data.columns:
    if train_data[col].dtypes == 'object':
        train_data[col] = la.fit_transform(train_data[col])
```

```python
train_data.info()
```

```python
train_data.head()
```

### Splitting the data into training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
train_data = train_data.values
```

```python
X,y = train_data[:,1:9], train_data[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=26)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.fit_transform(X_test)
```

## Building the model

```python
rescaledX_train
```

```python
from sklearn.linear_model import LogisticRegression

le = LogisticRegression()

le.fit(rescaledX_train, y_train)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel = 'linear' , probability = True, random_state = 42)

svm.fit(rescaledX_train, y_train)
```

```python
from keras import models
from keras import layers

network = models.Sequential()
network.add(layers.Dense(8, activation='relu', input_dim=8))
network.add(layers.Dense(64, activation='relu'))
network.add(layers.Dense(64, activation='relu'))
network.add(layers.Dense(64, activation='relu'))
network.add(layers.Dropout(0.4))
network.add(layers.Dense(2, activation='softmax')) 
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=150)
rf.fit(rescaledX_train, y_train)
```

```python
x_pred = rf.predict(rescaledX_test)

print("Accuracy of logistic regression classifier: ", rf.score(rescaledX_test, y_test))
```

## Making Prediction and evaluating model performance

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = le.predict(rescaledX_test)

print("Accuracy of logistic regression classifier: ", le.score(rescaledX_test, y_test))

confusion_matrix(y_pred,y_test)
```

```python
yy_pred = svm.predict(rescaledX_test)

print("Accuracy of SVM classifier: ", svm.score(rescaledX_test, y_test))

confusion_matrix(y_pred,y_test)
```

```python
network.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
```

```python
network.summary()
```

```python
network.fit(rescaledX_train, y_train, epochs=500, verbose=1)
```

```python
test_loss, test_acc = network.evaluate(rescaledX_test, y_test)
```
